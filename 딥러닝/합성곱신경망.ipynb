# 합성곱
- 입력데이터에 도장(필터)을 찍어서 유용한 특성만 드러나게 하는 것
- 각 특성에 뉴런의 가중치를 곱하여 절편을 더한 것

- 3개씩 묶어서 3개의 특성과 곱해져 1개의 출력을 만듬
    - 위의 과정을 특성이 끝날때까지 반복

image.png

- 가중치, 절편은 모든 합성곱이 같다.(첫번째 합성곱 = 두번째 합성곱)
- 단, 첫번째 특성, 마지막 특성의 영향이 적음 / 가운데 특성들의 과도한 사용
    
    => 정확도가 떨어질 가능성이 높음
    
    => **패팅**사용

image.png

- 합성곱 : n개의 배열 -> 1개의 가중치, 절편 -> (n - 가중치 + 1)개의 합성곱 출력
    - 가중치는 조정가능(하이퍼파라미터 : 홀수로 지정할 것)
    
    ex) 10개의 배열 -> 1개의 가중치, 절편 -> 8개(= 10 - 3 + 1)의 합성곱 출력
- 밀집층 : n개의 배열 -> n개의 가중치 -> 1개의 출력

- 경사하강법
    - 뉴런이 길게 늘어서 있고, 서로 조밀하게 연결
- 합성곱
    - 뉴런이 입력위를 이동하면서 출력을 만들기에 아래 그림처럼 표시 힘듬
    - 뉴런이라고 부르기 애매함.
    
    => 명칭 조정 : 필터 or 커널(핵심, 코어등등)

# 케라스 합성곱 층 테스트
- 라이브러리 :  keras.layers 패키지
    - Conv2D 클래스 : 입력데이터 위를 이동하면서 합성곱을 만듬

```python
from tensorflow import keras

keras.layers.Conv2D(10, kernel_size = (3,3), activation = 'relu')
#             필터(도장)의 갯수
# kernel_size : 커널 사이즈(3*3)
# 활성함수 : 렐루함수
```

## 특성 맵
- 활성화 함수 적용 후에 사용함(활성화 함수 생략가능)
- 일반적으로 특성 맵은 활성화 함수를 통과한 값을 나타냄
    - 합성곱에서는 활성화 출력이라는 표현을 잘 쓰지 않음
<br/><br/>
- 커널의 크기 : 하이퍼 하라미터로 대부분 3 * 3이나 5 * 5 크기를 권장함
- 케라스 API의 합성곱 층은 Dense층을 사용했던 자리에 대신 Conv2D 층을 넣음

## 패딩(padding)
- 배열의 주위를 가상의 원소로 채우는 것
    - 실제 값이 아니기 때문에 0으로 채움
    - 계산에 영향을 미치지는 않음
- 역활
    - 순전히 커널이 도장을 찍을 횟수를 늘려주는 것
    - 이유 : 모시리에 있는 특성의 반영 비율을 높이기 위해
- 세임패딩(padding="same")
    - 입력과 특성 맵의 크기를 동일하게 만들어 주기 위해 입력 주의를 0으로 패딩하는 것
    - 가장 많이 사용
- 밸리드 패딩(padding = 'valid') : 기본값
    - 패딩없이 순수한 입력 배열에서만 합성곱을 하여 특성 맵을 만듬(특성 맵의 크기가 줄어 들수 밖에 없음)
- 합성곱에서 패딩을 주로 사용하는 이유
    - 패딩x시, 각각의 모서리의 특성은 제대로 적용되지 않음

ex)
1. 입력 : (4,4)
2. 커널의 크기 : (3, 3)
3. 결과(특성맵) :  (2, 2)

- 추가사항 : 커널크기는 (3,3)으로 그대로 두고, 결과(특성맵)을 (4,4)로 반환하고 싶을때
    - 입력과 동일한 크기의 출력을 만들려면 마치 더 큰 입력에 합성곱인 척해야함
    
    => 빈 줄을 추가하여 (6, 6) 로 만들어야함

위 그림고 같이 (6,6) 크기이면 (3,3)크기의 합성곱을 했을 때의 출력의 크기는?




ex)
입력 : (4,4) 크기

1. 패딩x시 커널계산
    - 모서리 : 1회
    - 나머지 : 2회이상
    - 결론 : 가운데값이 두드러짐
        - 모서리 : 가운데 = 4 : 1

2. 패딩 o시
    - 모서리 : 4회
    - 나머지 : 6번이상
    - 결론 : 가운데값이 두드러짐
        - 모서리 : 가운데 = 9 : 4

3. 패딩 + 스트라이드 시
    - 모서리 : 가운데 = 1 : 1

```python
keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu', padding='same')
```

## 스트라이드(stride)
- 합성곱 연산은 좌우, 위아래 한칸씩이 기본값이지만 두칸씩 이동할 수도 있는 방법(특성맵의 크기가 작아짐)
- 스트라이드크기 = 풀링의 크기이므로 따로지정x (자동)

```python
keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu', padding="same", strides=1)
                                                                  # strides = 2로 사용할 일은 없음
```

## 풀링(pooling)
- 합성곱 층에서 만든 특성맵의 **가로세로 크기를 줄이는 역할**을 수행

    => 하지만 특성맵의 **개수**는 줄이지 않음
- 특성 맵에 커널 없이 필터를 적용하는 것과 비슷
    - 컬럼크기만큼 이동해서 다음 필터가 적용(겹치는 부분 없이 이동)
    - 폴링크기 = 스트라이드크기
- 폴링 종류 : 가중치가 없기때문에
    - 최대폴링 : 영역에서 가장 큰 값
    - 평균폴링 : 영역의 평균값
- 폴링층 : 합성곱 층과 구분되기에 따로 부름

ex)

입력 : (2,2,3)

출력 : (1,1,3) => 너비와 높이만 줄어듬

ex)

입력 : (4,4)

적용 : 최대(max) 풀링

출력 : (2,2) -> 이동시 2칸씩 이동(겹침없이 이동하기 위해)
<br/><br/>
<깊이 차원적용시>

입력 : (4,4, n)

폴링반복갯수 : 깊이차원, n개의 특성맵이 있다면 n번반복

출력 : 가중치가 없고, 폴링크기 = 스트라이드 크기(패딩x)

```python
keras.layers.MaxPooling2D(2) # 대부분 풀링의 크기는 2 (가로세로크기를 절반으로 줄임)
```

```python
keras.layers.MaxPooling2D(2, strides=2, padding="same") #최대풀링

keras.layers.AveragePooling2D(2, strides=2, padding="valid") #평균풀링
#대부분 최대풀링을 사용
# 평균풀링은 특성맵에 있는 중요한 정보를 평균을 내어 희석이 되어버림

#keypoint : 풀링은 가로 세로 방향만 진행(특성맵의 개수는 변하지 않고 그대로)
```

### 합성곱층-풀링층

이와 비슷한 경우는 합성곱 층-풀링틍 다음에 다시 합성곱 층이 올때

첫번째 합성곱 층의 필터 개수가 5개라고 가정하여 첫번째 풀링 층을 통과한 특성 맵의 크기가 (4,4,5)라고 했을때

두번째 합성곱 층에서 필터의 너비와 높이가 각각 3이라면 이 필터의 커널 크기는 (3,3,5)가됨

이유
- 입력의 깊이와 필터의 깊이가 같아야함(3 * 3 * 5 = 45개의 가중치를 곱하고 절편을 더함) => 합성곱의 결과 : 1개 출력되어야함

두번째 합성곱 층의 필터개수가 10개라면 만들어진 특성맵의 크기는 (2,2,10)이 됨
이렇게 합성곱 신경망은 너비와 높이는 점점 줄어들고 깊이는 점점 깊어지는 것이 특징
마지막에 출력층 전에 특성 맵을 모두 펼쳐서 밀집층의 입력으로 사용

합성곱 신경망에서 필터는 이미지에 있는 어떤 특징을 찾는다.
처음에는 간단한 기본적인 특징(직선, 곡선등)을 찾고 층이 깊어질수록 다양하고 구체적인 특지을 감지할 수 있도록 필터의 개수 늘림

또 어떤 특징이 이미의 어느 위치에 놓이더라도 쉽게 감지할 수 있도록 너비와 높이 차원을 점점 줄여가는 것이 합성곱 신경망의 목표

# 이미지 분류 실습
- 패션 MNIST를 활용
- 전처리
    - 데이터 스케일을 0~255 사이에서 0~1 사이로 바꾸고 훈련 세트와 검증 세트로 나눔
    - 차이점
        - 완전 연결 신경망에서는 입력 이미지를 밀집층에 연결하기 위해 일렬로 펼쳐야함(numby.reshape(), Faltten 클래스 사용)
        - 합성곱 신경망은 2차원 이미지르 그대로 사용하기에 필요x
- 입력이미지 : 깊이(채널)차원 존재
    - 흑백이미지의 경우, 채널차원이 없는 2차원 배열이지만 Conv2D층을 사용하기 위해 마지막에 채널( , , 1)을 추가
        - numpy.reshpe() 메서드를 이용하여 전체 배열 차원을 그대로 유지하면서 마지막에 차원을 간단히 추가

## 데이터

```python
import tensorflow as tf

tf.keras.utils.set_random_seed(42)

from tensorflow import keras
from sklearn.model_selection import train_test_split

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data() # 패션MNIST 데이터

train_scaled = train_input.reshape(-1, 28, 28, 1) / 255.0
# train_input(48000, 28, 28) -> train_scaled(48000, 28, 28, 1)로 4차원 변경

train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target, test_size=0.2, random_state=42)
#훈련데이터, 검증데이터 : 8:2
```

## 합성곱 신경망 생성
- 합성곱 층으로 이미지에서 특징을 감지한 후 밀집층으로 클래스에 따른 분류 확률을 계산함

```python
model = keras.Sequential()
model.add(keras.layers.Conv2D(32, kernel_size=3, activation='relu', padding="same", input_shape=(28,28,1)))
#           Conv2D층을 add로 추가
# 32                    : 합성곱 층으로 32개의 필터를 사용(경험으로 넣는 것)
# kernel_size = 3       : 커널크기(3,3), 차원채널은 입력값과동일
# activation='relu'     : 활성화함수 렐루함수이용
# padding="same"        : 세임 패딩이용
# input_shape=(28,28,1) : 앞에 변경한 이미지를 차원변경함
```

```python
model.summary()
# (None, 28, 28, 32) : none, 너비, 높이, 필터갯수
```

```python
model.add(keras.layers.MaxPooling2D(2))
# 풀링층 추가 (최대 풀링(2,2) 특성 맵의 크기는 절반으로 줄어듬)
# 합성곱 층에서 32개의 필터를 사용했기 때문에 특성 맵의 깊이는 32
# 최대풀링을 통과한 특성 맵의 크기는 (14,14,32) : 28 / 2 = 14
model.summary()
```

```python
# 이번 합성곱층은 세임패딩이용(입력의 가로,세로, 크기를 줄이지는 않음)
# 필터크기를 64로 올림
model.add(keras.layers.Conv2D(64, kernel_size=(3,3), activation='relu', padding="same"))
model.summary()
```

```python
# 폴링층에서 이 크기를 절반으로 줄임
# 64개의 필터를 사용했으니 특성맵의 크기는 (7, 7, 64) : 14 / 2 = 7
model.add(keras.layers.MaxPooling2D(2))
model.summary()
```

## 심층신경만생성

```python
# 3차원 특성맵을 일렬로 펼칠차례 -> 최종 : 10개의 뉴런을 가진 밀집층에서 활률을 계산
# 이번에는 특성맵을 일렬로 펼쳐서 바로 출력층에 전달하지 않고 중간에 하나의 밀집 은닉층을 하나 추가함
# Flatten -> Dense 은닉층 -> Dropout -> Dense 출력층
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(100, activation='relu'))
# 은닉층 : 100개의 뉴런사용, 활성화함수는 렐루이용
model.add(keras.layers.Dropout(0.4))
# 드룹아웃 : 은닉층의 과대적합 방지
model.add(keras.layers.Dense(10, activation='softmax'))
# 패션 MNIST는 클래스를 10개 분류하는 다중 분류 => softmax이용
model.summary()
```

## 모델층 확인
합성곱층과 풀링층의 효과를 확인가능
- 첫번째 합성곱 통과후 특성맵의 깊이 : 32
- 두번쨰 합성곱 통과후 특성맵의 깊이 : 64
- 반면 특성맵의 가로세로 크기
    - 첫번째, 두번째 풀링층 : 각각 절반으로 줄어듬
- 최종 : (7, 7, 64)

완전신경망에서 해던 것 처럼 모델 파라미터 개수
첫번째 합성곱층은 32개의 필터를 가지고 있고 크기가(3. 3. )깊이가 1이고 필터마다 1의 절편이 있음
3 * 3 * 1 * 32 + 32 = 320

두전째 합성곱층은 64개의 필터를 사용하게 크기가 (3, 3,) 깊이가 32이고 필터마다 1의 절편이 있음
3 * 3 * 32 * 64 + 64 = 18,496 (도장(*) * 깊이 * 만든깊이 + 절편)

Flatten 클래스의 (7,7,64) 크기의 특성맵을 1차원 배열로 펼치면 (3136,)의 크기의 배열이 됨 이를 100개의 뉴런과 완전히 연결해야함
7 * 7 * 64 = 3,136
3136 * 100 + 100 = 313,700

출력층의 모델 파라미터 개수 1010개
100 * 10 + 10 = 1,010

```python
# 버전차이확인(이전 : 흰색배경, 현 : 검은색배경)
keras.utils.plot_model(model)
```

## 모델 컴파일 & 훈련

```python
# 모델 컴파일
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 아담 최적화 훈련
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-cnn-model.keras', save_best_only=True)

# ModelCheckpoint 콜백 조기종료 -> 검증이 2번정도 상향시 조기종료
early_stopping_cb = keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)

# EarlyStoping 콜백 조기종료
history = model.fit(train_scaled, train_target, epochs=20,
                    validation_data= (val_scaled, val_target),
                    callbacks=[checkpoint_cb, early_stopping_cb])
```

### 시각화

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
```

## 성능평가(evaluate)

```python
# EarlyStopping 클래스에서 restore_best_weights=True로 지정해서
# 현재model객체가 최적의 모델 파라미터로 복원됨
# ModelCheckpoint 코백이 저장한 best-cnn-model.,keras 파일을 다시 읽을 필요가 없음

# 이번 세트에 대한 성능 평가 진행
model.evaluate(val_scaled, val_target)
# 결과 : fit() 메서드의 출력중 6번째와 비슷한 출력값
# EarlyStoppping 콜백이 model 객체의 최상의 모델 파라미터 잘 복원함
```

```python
test_scaled = test_input.reshape(-1, 28, 28,1) / 255.0
#훈련세트와 검증세트에서 했던 것처럼 0~1사이로 바꾸고 이미지크기 (28,28,1)로 변경
```

```python
model.evaluate(test_scaled, test_target)
```
